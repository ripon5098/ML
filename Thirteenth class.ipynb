```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex', size='size', style='smoker')
```

```python
tips['sex'][tips['sex'] == 'Male'].count()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker', col='smoker')
```

```python
sorted(tips['size'].unique())
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker', size='size')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker',col='smoker', size='size')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker',col='size', size='size', col_wrap=2)
```

```python
sns.scatterplot(x='day', y='tip', data=tips)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
sns.catplot(x='day', y='tip', data=tips)
```

```python
sns.catplot(x='day', y='tip', data=tips, kind='swarm')
```

```python
sns.catplot(x='day', y='tip', data=tips, kind='box')
```

```python
sns.catplot(x='day', y='tip', data=tips, kind='swarm', hue='sex')
```

```python
sns.catplot(x='day', y='tip', data=tips, kind='box', hue= 'sex')
```

```python
sns.catplot(x='day', y='tip', data=tips, kind='swarm', hue='smoker')
```
