```python
from abc import ABC, abstractmethod

class customer(ABC):
    @abstractmethod
    def get_name(self):
        pass

    @abstractmethod
    def set_name(self, name):
        pass

    @abstractmethod
    def get_age(self):
        pass

    @abstractmethod
    def set_age(self, age):
        pass

    @abstractmethod
    def get_nid(self):
        pass

    @abstractmethod
    def set_nid(self, nid):
        pass

    @abstractmethod
    def get_adress(self):
        pass

    @abstractmethod
    def set_adress(self, adress):
        pass

class customerImplementation(customer):

    def __init__(self, name, age, nid, adress):
        self.name = name
        self.age = age
        self.__nid = nid
        self.adress = adress


    def get_name(self):
        return self.name

    def set_name(self, name):
        self.name = name

    def get_age(self):
        return self.age

    def set_age(self, age):
        self.age = age

    def get_nid(self):
        return self.__nid

    def set_nid(self, nid):
        self.__nid = nid

    def get_adress(self):
        return self.adress

    def set_adress(self, adress):
        self.adress = adress

    def __str__(self):
        return "Name: "+self.name+" "+"\nAge: "+str(self.age)+" "+"\nNID: "+str(self.__nid)+" "+"\nAdress: "+" "+self.adress


customer = customerImplementation("Erha", 12 , 235654657, "ABB Colony")
print(customer)
```

```python
class savingsaccounts(customerImplementation):

    def __init__(self, name, age, nid, adress, accountnumber, deposit, withdraw):
        customerImplementation.__init__(self,name, age, nid, adress)
        self._accountnumber = accountnumber
        self.__openingbalance = 0.0
        self.__deposit = deposit
        self.__withdraw = withdraw
        
    def openingaccount(self):
        self.__openingbalance = int(input("Opening Transaction : "))
        if self.__openingbalance >= 500:
            print("Congratulations. You have opened your account.")
            
            def deposit(self):
                amount = int(input("Make a deposit : "))
                self.__openingbalance += amount

            def withdraw(self):
                amount = int(input("Withdrawal amount : "))
                if self.__openingbalance >= amount:
                    self.__openingbalance -= amount
                else:
                    print("Insufficient Balance")

            def info(self):
                return f"{self.name} has balance = {self.__openingbalance}"
                
            
        else:
            print("Account opening unsuccessful.")
            
    
            
c = savingsaccounts("Effat", 13, 345059608, "FGGFColony", "dshj56576")
print(c)
print(c._accountnumber)
c.openingaccount()
print(c.deposit())
print(c.withdraw())
print(c.info())
```
